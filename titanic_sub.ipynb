```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode().iloc[0])
```

```python
df['Embarked'].isna().sum()
```

```python
df['Cabin'] = df['Cabin'].fillna(df['Cabin'].mode().iloc[0])
```

```python
df['Deck'] = df['Cabin'].apply(lambda x: x[0] if pd.notnull(x) else 'M')
```

```python
df.info()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df = df.drop('Cabin', axis = 1)
```

```python
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
df = df.drop('Name', axis = 1)
```

```python
def title_replace(title):
    if title in ['Mlle', 'Ms']:
        return 'Miss'
    elif title == 'Mme':
        return 'Mrs'
    elif title in ['Don', 'Rev', 'Dr', 'Major', 'Lady', 'Sir', 'Col', 'Capt', 'Countess', 'Jonkheer', 'Dona']:
        return 'Rare'
    else:
        return title 
```

```python
df['Title'] = df['Title'].apply(title_replace)
```

```python
titles_m = {'Mr': 0, 'Mrs': 1, 'Miss': 2, 'Master': 3, 'Rare': 4}
df['Title'] = df['Title'].map(titles_m)
```

```python
def clean_ticket(ticket):
    if ticket.isdigit():
        return "Numeric"
    else:
        letters = ticket.split(' ')[0]
        letters = letters.replace('.', '').replace('/', '').upper()
        return letters

df['Ticket_Letters'] = df['Ticket'].apply(clean_ticket)

df['Ticket_Letters'].value_counts()
```

```python
def ticket_replace(ticket):
    main_tickets = ['A5', 'PC', 'Numeric', 'CA', 'SCPARIS', 'WC', 'SOTONOQ', 'STONO']
    
    if ticket in main_tickets:
        return ticket
    else:
        return 'Rare'
```

```python
df['Ticket_Letters'] = df['Ticket_Letters'].apply(ticket_replace)
```

```python
ticket_m = {'A5': 0, 'PC': 1, 'Rare': 2, 'Numeric': 3, 'CA': 4, 'SCPARIS': 5, 'WC':  6, 'SOTONOQ': 7, 'STONO': 8}
df['Ticket_Letters'] = df['Ticket_Letters'].map(ticket_m)
```

```python
df = df.drop('Ticket', axis = 1)
```

```python
deck_m = {'M': 0, 'C': 1, 'E': 2, 'G': 3, 'D': 4, 'A': 5, 'B': 6, 'F': 7, 'T': 8}
embarked_m = {
    'S': 0,
    'C': 1,
    'Q': 2
}
```

```python
df['Embarked'] = df['Embarked'].map(embarked_m)
df['Deck'] = df['Deck'].map(deck_m)
```
